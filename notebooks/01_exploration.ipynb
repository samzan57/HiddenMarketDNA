```python
# --- 1. Import des bibliothèques ---
import sys
import os

# Racine du projet (1 niveau au-dessus du notebook)
sys.path.append(os.path.abspath(".."))
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.data_loader import DataLoader
from src.returns import compute_log_returns

sns.set_style("whitegrid")
%matplotlib inline
```

```python
# --- 2. Configuration des ETF ---
ETF_LIST = ["XLK", "XLF", "XLE", "XLV", "XLI", "XLY", "XLP", "XLU"]
DATA_DIR = "data/raw"
```

```python
# --- 3. Chargement des prix ---
loader = DataLoader("../data/raw")
prices = loader.load_prices(ETF_LIST)

print("Prix ajustés (extrait) :")
display(prices.head())
```

```python
# --- 4. Statistiques descriptives ---
print("Statistiques descriptives des prix :")
display(prices.describe())

# Calcul des rendements logarithmiques
returns = compute_log_returns(prices)

print("Rendements log (extrait) :")
display(returns.head())

print("Statistiques descriptives des rendements :")
display(returns.describe())
```

```python
prices = loader.load_prices(ETF_LIST)

print(prices.dtypes)
print(prices.head())
```

```python
# --- 5. Corrélations ---
corr_matrix = returns.corr()
print("Matrice de corrélation :")
display(corr_matrix)

# Heatmap des corrélations
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Corrélation entre ETF sectoriels")
plt.show()
```

```python
# --- 6. Intuition sectorielle ---
# On peut regarder les rendements moyens et volatiles par ETF
mean_returns = returns.mean()
volatility = returns.std()

summary = pd.DataFrame({
    "Mean Return": mean_returns,
    "Volatility": volatility
}).sort_values(by="Volatility", ascending=False)

print("Résumé sectoriel :")
display(summary)

# Visualisation : moyenne vs volatilité
plt.figure(figsize=(8, 5))
plt.scatter(summary["Volatility"], summary["Mean Return"], s=100)

for i, ticker in enumerate(summary.index):
    plt.text(summary["Volatility"][i], summary["Mean Return"][i], ticker)

plt.xlabel("Volatilité (std)")
plt.ylabel("Rendement moyen")
plt.title("Intuition sectorielle : Rendement vs Volatilité")
plt.show()
```
